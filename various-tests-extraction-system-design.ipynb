```python
import certifi
import json
import pandas as pd
from urllib.request import urlopen
```

```python
pd.set_option("display.max_columns", 100)
```

### Load balance sheet info

```python
url_balance_sheet = "https://financialmodelingprep.com/api/v3/balance-sheet-statement/SSY?period=quarter&from=2018-12-31&to=2019-12-31"
response_balance_sheet = urlopen(url_balance_sheet, cafile=certifi.where())
data_balance_sheet = response_balance_sheet.read().decode("utf-8")
data_balance_sheet = json.loads(data_balance_sheet)
```

```python
balance_sheet_df = pd.DataFrame(data_balance_sheet)
balance_sheet_df = balance_sheet_df.set_index("date")
balance_sheet_df
```

```python
balance_sheet_df.tail()
```

### Load income stmt info

```python
url_income_stmt = "https://financialmodelingprep.com/api/v3/income-statement/SSY?period=quarter&from=2018-12-31&to=2019-12-31"
response_income_stmt = urlopen(url_income_stmt, cafile=certifi.where())
data_income_stmt = response_income_stmt.read().decode("utf-8")
data_income_stmt = json.loads(data_income_stmt)
```

```python
income_stmt_df = pd.DataFrame(data_income_stmt)
income_stmt_df = income_stmt_df.set_index("date")
income_stmt_df
```

```python
income_stmt_df.tail()
```

### Manipulate balance sheet & income stmt columns

```python
# these columns appear both in the balance sheet and in the income statement
income_stmt_cols_to_drop = [
    "symbol",
    "reportedCurrency",
    "cik",
    "fillingDate",
    "acceptedDate",
    "calendarYear",
    "period",
    "finalLink",
    "link",
]
```

```python
income_stmt_df = income_stmt_df.drop(columns=income_stmt_cols_to_drop, axis=1)
```

```python
balance_sheet_df.columns = [col+"_bs" for col in balance_sheet_df.columns]
income_stmt_df.columns = [col+"_is" for col in income_stmt_df.columns]
```

```python
balance_sheet_df.head()
```

### Concatenate sources

```python
type(engine)
```

```python
financial_info = pd.concat([balance_sheet_df, income_stmt_df], axis=1)
```

```python
financial_info = financial_info.reset_index()
financial_info.head()
```
