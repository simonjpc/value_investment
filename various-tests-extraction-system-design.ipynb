```python
import certifi
import json
import pandas as pd
from urllib.request import urlopen
```

```python
url = "https://financialmodelingprep.com/api/v3/balance-sheet-statement/SSY?from=2018-12-31&to=2019-12-31"
response_balance_sheet = urlopen(url, cafile=certifi.where())
data_balance_sheet = response_balance_sheet.read().decode("utf-8")
data_balance_sheet = json.loads(data_balance_sheet)
```

```python
pd.DataFrame(data_balance_sheet)
```
